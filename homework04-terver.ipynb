### 1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800]. Найдите ее среднее значение и дисперсию.

```python
a = 200
b = 800
MA = (a+b)/2
DA = ((b-a)**2)/12
print(f'Среднее значение величины А: {MA}')
print(f'Дисперсия величины А: {DA}')
```

### 2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.

```python
a = 0.5
DB = 0.2
b =  (DB*12)**(1/2) + a
MB = (a+b)/2
print('Правая граниа:', round(b, 4))
print('Среднее значение:', round(MB, 4))
```

### 4. Рост взрослого населения города X имеет нормальное распределение. Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см. Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:

**_1) больше 182 см_**

**_2) больше 190 см_**

**_3) от 166 см до 190 см_**

**_4) от 166 см до 182 см_**

**_5) от 158 см до 190 см_**

**_6) не выше 150 см или не ниже 190 см_**

**_7) не выше 150 см или не ниже 198 см_**

**_8) ниже 166 см_**

```python
def z_value(x, mu, sigma):
    return (x-mu)/sigma

sigma = 8
mu = 174

x = 182
p1 = 100 - 84.13
print(f'1) Рост больше {x}: при Z = {z_value(x, mu, sigma)}, p1 = {round(p1, 2)}%')

x = 190
p2 = 100 - 97.72
print(f'2) Рост больше {x}: при Z = {z_value(x, mu, sigma)}, p2 = {round(p2, 2)}%')

x1 = 166
x2 = 190
p3 = 97.72 - 15.87
print(f'3) Рост от {x1} до {x2}: при Z = ({z_value(x1, mu, sigma)}, {z_value(x2, mu, sigma)}), p3 = {round(p3, 2)}%')

x1 = 166
x2 = 182
p4 = 84.13 - 15.87
print(f'4) Рост от {x1} до {x2}: при Z = ({z_value(x1, mu, sigma)}, {z_value(x2, mu, sigma)}), p4 = {round(p4, 2)}%')

x1 = 158
x2 = 190
p5 = 97.72 - 2.28
print(f'5) Рост от {x1} до {x2}: при Z = ({z_value(x1, mu, sigma)}, {z_value(x2, mu, sigma)}), p5 = {round(p5, 2)}%')

x1 = 150
x2 = 190
p6 = 0.13 + 100 - 97.72
print(f'6) Рост не выше {x1} или не ниже {x2}: при Z <= {z_value(x1, mu, sigma)}, >= {z_value(x2, mu, sigma)}), p6 = {round(p6, 2)}%')

x1 = 150
x2 = 198
p7 = 0.13 + 100 - 99.87
print(f'7) Рост не выше {x1} или не ниже {x2}: при Z <= {z_value(x1, mu, sigma)}, >= {z_value(x2, mu, sigma)}), p7 = {round(p7, 2)}%')

x = 166
p8 = 15.87
print(f'8) Рост ниже {x}: при Z = {z_value(x, mu, sigma)}, p8 = {round(p8, 2)}%')
```

### 5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

```python
MX = 178
DX = 25
X = 190
sigma = DX ** (1/2)
Z = (X-MX)/sigma
print('Отклонение от математического ожидания:', Z)
```
