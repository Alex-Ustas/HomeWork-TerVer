### 1. Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks)
Используя математические операции, посчитать коэффициенты линейной регрессии, приняв за X заработную плату (то есть, zp - признак), а за y - значения скорингового балла (то есть, ks - целевая переменная). Произвести расчет как с использованием intercept, так и без.

```python
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])

plt.scatter(zp,ks)
plt.xlabel('Величина заработной платы')
plt.ylabel('Поведенческий кредитный скоринг')
plt.show()
```

```python
n = len(ks)
b1 = (np.mean(zp * ks) - np.mean(zp) * np.mean(ks)) / (np.mean(zp ** 2) - np.mean(zp) ** 2)
b0 = np.mean(ks) - b1 * np.mean(zp)

y_pred = b0 + b1 * zp

plt.scatter(zp,ks)
plt.plot(zp, y_pred, c='g')
plt.xlabel('Величина заработной платы')
plt.ylabel('Поведенческий кредитный скоринг')
plt.show()
```

```python
zp1 = zp.reshape(1, n)
ks1 = ks.reshape(1, n)

b1 = np.dot(np.dot(np.linalg.inv(np.dot(zp1, zp1.T)), zp1), ks1.T)[0][0]

y_pred1 = b1 * zp

plt.scatter(zp,ks)
plt.plot(zp, y_pred, 'g', label = 'intercept')
plt.plot(zp, y_pred1, 'r', label = 'без intercept')
plt.legend()
plt.show()
```

### 2. Посчитать коэффициент линейной регрессии при заработной плате (zp), используя градиентный спуск (без intercept).

```python
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
```

```python
def _mse(b, x, y):
    return np.sum((b * x - y) ** 2) / len(x)

def _mse_p(b, x, y):
    return (2 / len(x)) * np.sum((b * x - y) * x)
```

```python
# Параметр скорости обучения
alpha = 1e-06

b = 0.1
mse_min = _mse(b,zp,ks)
i_min =  1
b_min = b
for i in range(10000):
    b -= alpha * _mse_p(b,zp,ks)
    if i%100==0:
        print(f'Итерация #{i}, b={b}, mse={_mse(b, zp, ks)}')
    if _mse(b,zp,ks)>mse_min:
        print(f'Интерация #{i_min}, b={b_min}, mse={mse_min}, \n-- Достигнут минимум --')
        break
    else:
        mse_min = _mse(b,zp,ks)
        i_min = i
        b_min = b

print(b_min)
```

### 3. Произвести вычисления как в пункте 2, но с вычислением intercept. Учесть, что изменение коэффициентов должно производиться на каждом шаге одновременно (то есть изменение одного коэффициента не должно влиять на изменение другого во время одной итерации).

```python
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
```

```python
# Функиция потерь
def _mse_ab(a,b,x,y):
    return np.sum(((a + b * x)- y) ** 2) / len(x)

# Функиция потерь по a
def _mse_pa(a,b,x,y):
    return 2*np.sum((a + b * x)- y) / len(x)

# Функиция потерь по b
def _mse_pb(a,b,x,y):
    return 2*np.sum(((a + b * x)- y) * x) / len(x)
```

```python
#Скорость обучения
alpha = 5e-05

b=0.1
a=0.1
mseab_min=_mse_ab(a,b,zp,ks)
i_min=1
b_min=b
a_min=a
   
for i in range(1000000):
    a-=alpha*_mse_pa(a,b,zp,ks)
    b-=alpha*_mse_pb(a,b,zp,ks)
    if i%50000==0:
        print(f'Итерация #{i}, a={a}, b={b}, mse={_mse_ab(a, b, zp,ks)}')
    if _mse_ab(a, b,zp,ks)>mseab_min:
        print(f'Итерация #{i_min}, a={a_min}, b={b_min}, mse={mseab_min},\n-- Достигнут минимум --')
        break
    else:
        mseab_min=_mse_ab(a, b,zp,ks)
        i_min=i
        b_min=b
        a_min=a
print(f'a={a_min}\nb={b_min}')
```

```python
plt.scatter(zp, ks)
plt.plot(zp,a_min + b_min * zp, c='r')
plt.xlabel('Величина заработанной платы')
plt.ylabel('Поведенческий кредитный скоринг')
plt.show()
```
